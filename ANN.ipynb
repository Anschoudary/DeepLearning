# **Deep Learning**

    Deep learning is a brach of machine learning that mimics Human Brain and find the complex patterns in data using neural networks. It is uses in many advance algorithms nowadays.

## Examples:

    Images processing
    Object classification

```python
# Required Libraries

import tensorflow as tf
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from tensorflow import keras
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
```

## **Dataset**

    We will use make_moons dataset from sklearn.datasets

```python
x, y = make_moons(n_samples=10000, random_state=2)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

```python
df = pd.DataFrame(x, y)
df.head()
```

## **Building Model**

    We will make a neural network having 4 layers in total (1 input, 2 hidden and 1 output layer)

```python
def build_model():
  model = keras.Sequential([
      keras.layers.Flatten(input_shape=(2,)),
      keras.layers.Dense(30, activation='relu'),
      keras.layers.Dense(30, activation='relu'),
      keras.layers.Dense(1, activation='sigmoid')],
      name = 'NN_MLP_Model')
  model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

  return model
```

```python
model = build_model()
model.summary()
```

## **Model Training**

```python
history = model.fit(x_train, y_train, epochs=50)
```

```python
accuracy = history.history['accuracy']
loss = history.history['loss']

print(accuracy)
print(loss)
```

## **Accuracy & Loss**

```python
plt.plot(accuracy)
plt.plot(loss)
plt.legend(['accuracy', 'loss'])
plt.show()
```

```python
from sklearn.metrics import accuracy_score

# Calculate predictions
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5).astype(int)

accuracy_score = accuracy_score(y_test, y_pred)

print(accuracy_score)
```

## **Heatmap**

```python
# Generate confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
import seaborn as sns
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```
